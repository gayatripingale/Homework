```python
# 2-1/api/v1.0/precipitation querys. 

@app.route("/api/v1.0/precipitation")

def precipitation():
    
    # query the database for dates and tobs for last year
    prcp_result = session.query(Measurement.date,Measurement.tobs).\
    filter(Measurement.date > '2017-01-01').all()
    
    # 2-2 Convert the query results to a Dictionary using date as the key and tobs as the value.
    # We will do that by creating a dictioonary from row data and append to the list "all_prcp"
    all_prcp = []
    for prcp in prcp_result:
        prcp_dict = {}
        prcp_dict["Date"] = Measurement.date
        prcp_dict["tobs"] = Measurement.tobs
        all_prcp.append(prcp_dict)
        # 2-3 Return jsonify list
    return jsonify(all_prcp)     
```
